```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
N = 1000
X = np.random.rand(N)*2-1
```

```python
beta = 1.0
burnin_steps = 1000
record_steps = 10000
dt = 0.0001
sigma = sqrt(2.0 * dt / beta)
```

```python
for i in range(burnin_steps):
    X += np.random.randn(N) * sigma
    inbound = np.abs(X) < 1.0
    X = np.where(inbound, X, X/X**2)
Y = X
for i in range(record_steps):
    X += np.random.randn(N) * sigma
    inbound = np.abs(X) < 1.0
    X = np.where(inbound, X, X/X**2)
    Y = np.concatenate((Y, X))
```

```python
(h, bins) = np.histogram(Y, range=(-1, 1), bins=40)
plt.stairs(h, bins)
plt.show()
```
